```python
spec_est_scores=[]
top_scorers=[]
scorer='f1_micro'#'roc_auc_score_micro'
est_bigmean_describe={}

for est in ests:
    est_df=scor_df.loc[(slice(None),est),'scorer:'+scorer]
    #cvmean=est_df.mean(axis=1)
    cvmean=est_df.mean(axis=1,level='rep_idx').mean(axis=1)
    bigmean=cvmean.mean(axis=0,level='species')
    est_bigmean_describe[est]=bigmean.describe(percentiles=[.025,.975])
    top_scorers.append([bigmean.index[i] for i in np.argsort(bigmean)[-5:]][::-1])
    #print(bigmean.iloc[np.where(bigmean==1.)])
    spec_est_scores.append(bigmean.to_numpy())
    
```

```python
pd.DataFrame(top_scorers,index=ests)
```
